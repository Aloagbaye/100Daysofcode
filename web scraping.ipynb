```python
chem=pd.DataFrame()
for number in range(800,810):
    try:
        url="https://www.osha.gov/chemicaldata/"+str(number)
        df= pd.read_html(url)
        df=df[1]
        df.columns=["Physical Properties", str(number)+" Physical Properties.1",str(number)+" Physical Properties.2",str(number)+" Physical Properties.3"]
        if len(chem)==0:
            chem = df
        else:
            chem=chem.merge(df,left_on="Physical Properties",right_on="Physical Properties",how="left") 
    except Exception as e:
        print(url)
        print("no chemical for this code.")
```

```python
chem = chem.T
```

```python
chem.to_csv("osha numbers.csv")
```
