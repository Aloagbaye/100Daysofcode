```python
import pandas as pd
from string import ascii_lowercase
```

```python
chem=pd.DataFrame()
for c in ascii_lowercase:
    try:
        url="https://www.osha.gov/chemicaldatabase?chemical_first_letter_filter="+str(c)
        df= pd.read_html(url)
        df=df[0]
        if len(chem)==0:
            chem = df
        else:
            chem = pd.concat([chem,df])
    except Exception as e:
        print(url)
        print("no chemical for this code.")
```

```python
chem.reset_index(drop=True)
```

```python
chem.to_csv("chemicals.csv")
```
